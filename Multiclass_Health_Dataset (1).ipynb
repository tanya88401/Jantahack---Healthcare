```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x,y, test_size = 0.3, random_state = 37)

#display(X_train, X_test, Y_train, Y_test)
```

```python
X_train.isnull().sum()
```

```python
print("Length of X_train", len(X_train), 'length of Y_train', len(Y_train))
print('Length of X_test', len(X_test), 'Length of Y_test', len(Y_test))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier

model = LogisticRegression()

ovr = OneVsRestClassifier(model)
# fit model
ovr.fit(X_train, Y_train)
# make predictions
yhat = ovr.predict(X_train)
```

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix

print('Logistic Regression Accuracy',metrics.accuracy_score(yhat,Y_train))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

grad_boost_model = GradientBoostingClassifier()

grad_boost_model.fit(X_train,Y_train)
```

```python
y_pred_test = grad_boost_model.predict(X_test)
```

```python
y_pred_test
```

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
print('Gradient boost accuracy',metrics.accuracy_score(y_pred_test,Y_test))
```
